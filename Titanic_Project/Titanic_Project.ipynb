## Surviving the Titanic

<font color="blue">What was the chance of surviving of someone who paid more than 100 GBP for their ticket?</font>

```python
chance=len(titanic[(titanic["Fare"] > 100) & (titanic["Survived"] == 1)]) / len(titanic[titanic["Fare"] > 100])
chance= chance *100
print(f"{chance:.{2}f}%")
```

<font color="blue">We make a function that given a fare value of F calculates what percentage of people survived who had paid F or less. We use this function to plot the rate of survival against the fare paid for the fairs up until the 0.75 quantile (75% percentile)</font>

```python
def survived_fare(fare):
  chance=len(titanic[(titanic["Fare"] <= fare) & (titanic["Survived"] == 1)]) / len(titanic[titanic["Fare"] <= fare])
  chance= chance *100
  return chance

fare_75 = titanic["Fare"].quantile(0.75)
fare_values = range(int(titanic["Fare"].min()), int(fare_75) + 1)
survival_rates = [survived_fare(fare) for fare in fare_values]

plt.plot(fare_values, survival_rates)
plt.title("Survival Rate vs Fare Paid (Up to 75% Percentile)")
plt.xlabel("Fare")
plt.ylabel("Survival Rate (%)")
plt.grid(True)
```

## Modelling our data

---

We can consider outliers data with less than 9 fare.

```python
# Your Code Here
filtered_fare_values = range(9, int(fare_75) + 1)
filtered_survival_rates = [survived_fare(fare) for fare in filtered_fare_values]
plt.plot(filtered_fare_values, filtered_survival_rates)
plt.title("Filtered Survival Rate vs Fare Paid (Up to 75% Percentile)")
plt.xlabel("Filtered Fare")
plt.ylabel("Survival Rate (%)")
plt.grid(True)
```
